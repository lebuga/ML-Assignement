```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
plt.style.use('ggplot')
%matplotlib inline
```

```python
#Also check for Outliers in all the features
#an outlier is a data point that differs significantly from other observations,
#Box Plots
base_color = sb.color_palette()[0]
sb.boxplot(data = dataset,  y ='chol', color = base_color)
plt.xticks(rotation =15);
```

```python
def removal_box_plot(dataset, column, threshold):
    #sb.boxplot(dataset[column])
   # plt.title(f'Original Box Plot of {column}')
   # plt.show()

    removed_outliers = dataset[dataset[column] <= threshold]

    sb.boxplot(removed_outliers[column])
    plt.title(f'Box Plot without Outliers of {column}')
    plt.show()
    return removed_outliers


threshold_value = 380

no_outliers = removal_box_plot(dataset, 'chol', threshold_value)
```

```python
#Loading the Libraries
#Pre-requisite package imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
plt.style.use('ggplot')
%matplotlib inline
```

```python
#This will now get the top 10 most occuring persons in age and plot a bar graph
dataset['age'].value_counts()\
.head(10)\
.plot(kind ='bar', title ='Top 10 persons with most Age occurence')
plt.xlabel('Age occurence')
plt.ylabel('counts')
```

```python
#We could also vizualize cholesterol levels and occurence by making a plot, using histogram
#The histogram shows us in different bins the count of each value
ax =dataset['chol'].plot(kind ='hist',
                         bins =20,
                         title ='Choleterol Occurence')#you can add different bin sizes, this makes it cleaar to visualize
ax.set_xlabel('Cholesterol occurence')
ax.set_ylabel('Frequency of occurence')
#The Cholesterol distribution implies that between 200 and 300, there is high cholesterol occurence
#Also on the extreme far right, there are isolated cases of choleterol occurence
```

```python
#We can as well use a KDE to plot the cholesterol occurence
ax =dataset['chol'].plot(kind ='kde',
                         title ='Choleterol Occurence')
ax.set_xlabel('Cholesterol occurence')
ax.set_ylabel('Frequency of occurence')
#This kde makes the distrubution clear  and very similar to the histogram
```

```python
#Understanding the relationship between the features
#Having undertood the different features and their distributions of a given column in a dataset using plots
#Scatter PLot
#HeatMap Plot Correlation
#PairPlot
#GroupBy Comparisons

#Now we want to find out how did the different features relate to each other in the dataset
#we have to compare the different features in our dataset
```

```python
#Now lets plot a relationship between our age and cholesterol levels
dataset.plot(kind ='scatter', x='chol', y = 'age', title = 'Age Vs Cholesterol levels')
#plt.plot([100,500], [50,300])
#the plot implies btn 40 and 70, there are high occurences of cholesteral levels
plt.show()#helps to remove the matlib objects created by pandas
```

```python
#We can also use seabom to explore the relationshio betwen two features
#Establish relations betwen resting blood pressure and maximum heart rate
#We can add another feature to be the color (ie the hue = chol)
#We can see we now have different colors based on the choleterol levels ie color the csatter plot based on another variable
sb.scatterplot(x='thalach',
               y ='trestbps',
               hue = 'chol',
               data = dataset)
```

```python
#How to compare more that two variables in our dataset
sb.pairplot(dataset, vars =['age', 'chol', 'trestbps', 'thalach'], hue ='oldpeak')
plt.show()
```

```python
#We can also look at correlation to compare relations between features
#we drop any known null values and then do a correlation
#s seen, some fetures have negative correlatins and others have big correlations
#eg. a positive correlation betwen age and chol implies an increase in age leads to increase in cholesterol(weak correlations)
#nagative correlation btn age and maximum bps implies an increase in age leads to decrease in bp but the correlation is weak one (<-0.3>)
dataset_corr =dataset[['age', 'chol', 'trestbps', 'thalach']].dropna().corr()
dataset_corr
```

```python
#Another way to perform correlation is by using the Seabon's HeatMap method
#visualize how correlated the values are to each other
#also add annotations to see the raw values
#each value will have a perfect correlation with itself
sb.heatmap(dataset_corr, annot =True)
```

```python
#Asking a question about the data
#This is the final step of the Exploratory Data Analysis Process
#The we try to answer the question about the data using a plot or statistics
```

```python
#Qn. Which age bracket has the highest cholesterol levels (minimum of 200 cholesterol and above 5 count)?
#dataset['age'].value_counts()
#As you you can see from the statistics below, the age bracket of 41-67 has the highest average chols levesl of above 200 and most occuring
dataset.groupby('age')['chol']\
       .agg(['mean', 'count'])\
       .query('mean >=200 and count >=5')\
       .sort_values('mean')
```
